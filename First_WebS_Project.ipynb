```
from kora.selenium import wd
from bs4 import BeautifulSoup
import re

class Webpage:
     def __init__(self,url_company):
       self.url_company = url_company
 
       print(self.url_company)

     def scraper(self):
       wd.get(self.url_company)
       pagina = BeautifulSoup(wd.page_source)
       lista_empleos = pagina.find('div',class_='col-sm-12 mt15 listado-avisos')
       vacantes = lista_empleos.find_all('h2',class_='titulo-aviso')
       link_vacantes = lista_empleos.find_all('a')

       vacantes = [str(x) for x in vacantes] #convierte de objeto webpage a string
       link_vacantes = [str(x) for x in link_vacantes] #convierte de objeto webpage a string

       list_vacantes = []
       list_enlaces = []
       for vacante in vacantes:
         titulo = re.search("""title=(["'])([A-Z]\w.+)(["'])""",vacante)
         #print(titulo.group(2))
         list_vacantes.append(titulo.group(2))
       for link in link_vacantes:
         enlace = re.search("""href=(["'])(/empleos/[a-zA-Z]\w.+)(["'])""",link)
         if enlace is None:
           continue
         #print(enlace.group(2))
         list_enlaces.append(enlace.group(2))
       print(list_vacantes)
       print(list_enlaces)
       for i in range(1,len(list_vacantes)):
         puesto = list_vacantes[i]
         enlace = "https://www.bumeran.com.pe"+list_enlaces[i]

         print("Puesto: {} \nEnlace: {}".format(puesto,enlace))
       
     def mailsender(self):
       pass
       
     #  source = requests.get(self.url_company).text
     #  soup = BeautifulSoup(source,'lxml')
     #  articulo = soup.find('header')
     #  print(articulo)





def main():
  softys = Webpage("https://www.bumeran.com.pe/perfiles/empresa_productos-tissue-del-peru-s.a.c-softys-peru_10131768.html")
  softys.scraper()
  molitalia = Webpage("https://www.bumeran.com.pe/perfiles/empresa_molitalia-s.a._2233720.html")
  molitalia.scraper()

if __name__ == "__main__":
  main()
```
